```python
import re
```

```python
texto = "UE=12 c-rnti=0x4601 RSSI=-73 dBm"
```

```python
m = re.search(r"UE=(\d+).*c-rnti=(0x[0-9a-fA-F]+)", texto)
if m:
    print("UE:", m.group(1))
    print("C-RNTI:", m.group(2))
```

```python
texto = '''
Puedes contactar conmigo en los siguientes correos:

- personal: alvaro.duran@example.com
- trabajo: a.duran+dev@company-tech.io
- alternativo: user_test123@mail.co.uk
'''
```

```python
p = re.compile(r"([\w\.+]+)@([a-zA-Z_-]+\.[a-zA-Z\.-]+)")
```

```python
p
```

```python
m = re.search(p, texto)
if m:
    print("Nombre:", m.group(1))
    print("Dominio:", m.group(2))
```

```python
m = re.search(p, texto)
if m:
    print("Nombre, Dominio:", m.groups())
```

```python
m = re.search(p, texto)
if m:
    print(f"Nombre: {m.group(1)}\nDominio: {m.group(1)}")
```
